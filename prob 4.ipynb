**Πρόβλημα 4  Crude Method**


Untitled Notebook.png

```python
import numpy as np
import matplotlib.pyplot as plt

def u():
  return np.random.rand()

def f(x,y):
  return (20/13.)*(x+y)

N=0
V=2/3
Σf=0
Σf2=0

for i in range(1000):
  x=u()
  y=u()
  p=f(x,y)
  if x>=0 and y<=1 and  x**2<=y:
    Σf+=p
    Σf2+=p**2
    N+=1

mean=Σf/N
I=V*mean

sf2=(Σf2/N)-mean**2
δI=(V/np.sqrt(N))*np.sqrt(sf2)


print('I=%.3f +/- %.3f' %(I,δI))
```

```python
import numpy as np
import matplotlib.pyplot as plt


def p(x,y,z):
  return (12/31.)*(x**2+y*z)

def un(a,b):
  return (b-a)*np.random.rand() + a

N=0
V=1
Σf=0
Σf2=0

for i in range(1000):
  x=un(0,1)
  y=un(1,2)
  z=un(1,2)
  S=p(x,y,z)
  if 0<=x<=1 and 1<=y<=2 and  1<=z<=2:
    Σf+=S
    Σf2+=S**2
    N+=1

mean=Σf/N
I=V*mean

sf2=(Σf2/N)-mean**2
δI=(V/np.sqrt(N))*np.sqrt(sf2)


print('I=%.3f +/- %.3f' %(I,δI))
```

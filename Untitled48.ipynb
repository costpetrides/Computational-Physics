**Random walk theory**



```python
import numpy as np
import matplotlib.pyplot as plt

def step():
  res= 0
  if np.random.rand() < 0.5: 
    res=-1
  else:
    res=1
  return res


x=[]
for i in range(100):      # i atomo 
  thesi=[0]              # Arxiki thesi
  for n in range(100):    # vimata
    thesi+= [step()  + thesi[-1]]
  
  x+=[thesi]

x=np.array(x)      # Make x-list an array !
print(x) 

plt.plot(x[1])
plt.show()

plt.hist(x[:,-1], bin=10)
plt.show()
```

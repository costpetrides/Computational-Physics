```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
v=[]
file=open("data.txt","r")

N=0
for x in file:
  v.append(float(x))
  N+=1


mean=np.sum(v)/N
sd=np.sum((v-mean)**2)/N

print('N= ',N)
print('The mean value is: ',mean) 
print('The Variance: ',sd)

bins=range(1,8)
plt.hist(v, bins, ec = 'black')
plt.xlabel('Xj')
plt.ylabel('Frequency')

print(v)
```

```python
import matplotlib.pyplot as plt
import numpy as np

A = (20, 35, 30, 35, 27)        # Set of A data
B = (25, 32, 34, 20, 25)        # Set of B data
names= ['January', 'February', 'March', 'April', 'May']     # Set of names for x asis 

fig = plt.figure(figsize=(5,3), dpi=900)
left, bottom, width, height = 0.1, 0.2, 0.8, 0.6
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.3
ticks = np.arange(len(names))    
ax.bar(ticks, A, width, label='2018-2019')
ax.bar(ticks + width, B, width, align="center", label='2020-2021', color='skyblue')

ax.set_ylabel('CO2')
ax.set_title('Months')
ax.set_xticks(ticks + width/2)
ax.set_xticklabels(names)

ax.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
v=[]
file=open("data.txt","r")

N=0
for x in file:
  v.append(float(x))
  N+=1


mean=np.sum(v)/N
sd=np.sum((v-mean)**2)/N

print('N= ',N)
print('The mean value is: ',mean) 
print('The Variance: ',sd)

bins=range(1,8)
plt.hist(v, bins, ec = 'black')
plt.xlabel('Xj')
plt.ylabel('Frequency')

print(v)
```

```python
import matplotlib.pyplot as plt
import numpy as np

S21 = (7,4,4,4,6,9,5,5,6,5,7,9)        # Set of A data
S20 = (16, 18, 14, 4, 4, 6, 4, 4, 5, 6, 6, 8)        # Set of B data
S19 = ( 0, 0, 5, 6, 8, 10, 8, 6, 6, 11, 10, 10)    
names= ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN','JUL', 'AUG', 'SEP', 'OCT','NOV','DEC']     # Set of names for x asis 

fig = plt.figure(figsize=(10,3), dpi=200)
left, bottom, width, height = 0.1, 0.2, 0.8, 0.6
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.3
ticks = np.arange(len(names))    
ax.bar(ticks, S21, width, label='2021')
ax.bar(ticks + width, S20, width, align="center", label='2020', color='skyblue')
ax.bar(ticks + 2*width, S19, width, align="center", label='2019', color='lightblue')


ax.set_ylabel('S02')
ax.set_title('Months')
ax.set_xticks(ticks + width/2)
ax.set_xticklabels(names)

ax.legend(loc='best')
plt.show()
```

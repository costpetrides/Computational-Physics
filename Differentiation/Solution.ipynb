```python
def E(r,z):
  return z/(pow(r*r + z*z,1.5)) + (1-z)/(pow(r*r + (1-z)*(1-z),1.5))

z=np.linspace(0.1,0.9,81)
r=[0.1, 0.3, 0.41, 1]


for ri in range(len(r)):
  plt.plot(z,E(r[ri],z))
  plt.xlabel("z (h)")
  plt.ylabel(r"E(z) ($Q/4\pi\epsilon_0 h^2 $)")
  plt.title("r =" + str(r[ri]) + " h")
  plt.show()
```

# ii.

```python
def d2E(r,z):
  return 15*z*z*z/(pow(r*r + z*z,3.5)) - 9*z/(pow(r*r + z*z,2.5)) + 15*(1-z)*(1-z)*(1-z)/(pow(r*r+(1-z)*(1-z),3.5)) - 9*(1-z)/(pow(r*r+(1-z)*(1-z),2.5))

El=[]
d2El2=[]
d2E_NG2l=[]
error2=[]
error2l=[]
z2=np.linspace(0.1,0.87,78)

for ri in range(len(r)):
  d2Ei2=[d2E(r[ri],z2[zi]) for zi in range(len(z2))]
  d2El2.append(d2Ei2)
  d2Ei2=[]

for ri in range(len(r)):
  Ei=[E(r[ri],z[zi]) for zi in range(len(z))]
  El.append(Ei)
  Ei=[]

for ri in range(len(r)):
  d2E_NG2=[(El[ri][i+2]-2*El[ri][i+1]+El[ri][i])/(0.01*0.01) for i in range(len(z2))]
  d2E_NG2l.append(d2E_NG2)
  d2E_NG2=[]

for ri in range(len(r)):
  error2=[abs((d2E_NG2l[ri][pos]-d2El2[ri][pos])/d2El2[ri][pos])*100 for pos in range(len(d2El2[ri]))]
  error2l.append(error2)
  error2=[]

for nni in range(len(r)):
  print(" --------------")
  print("| For r=",r[nni],"h |")
  print(" --------------")
  print("z(h)\t\tN-G p_2\t\t\t d^2E/dz^2 \t\t error \t")
  print()
  for x,w,o,e in zip(z2,d2E_NG2l[nni],d2El2[nni],error2l[nni]):
    print(round(x,10),"\t\t",round(w,5),"\t\t",round(o,4),"\t\t",round(e,5),"%")
    print()
    print()
    plt.plot(z2,d2E_NG2l[nni],label="Newton-Gregory $p_2$ at $x_0$")
    plt.plot(z2,d2El2[nni],label="Analytical")
    plt.xlabel("z (h)")
    plt.legend(loc="center")
    plt.show()
```

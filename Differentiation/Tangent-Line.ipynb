```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2 +4

def df(x):
  h=0.009
  return (f(x+h)-f(x))/h

xx=np.linspace(-10,10,10**3)
yy=[f(x) for x in xx]



def dy(x):
  m=df(4)
  b=f(4)
  return m*(x-4)+b

dF=[dy(x) for x in xx]

plt.plot(xx,yy,label='f(x)')
plt.plot(xx,dF,label='df(4)')
plt.legend()
plt.show()
```

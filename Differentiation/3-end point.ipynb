Find the derivative of $y'(x)=x.e^{x}$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*exp(x)

h=0.01
x=2
df =(-f(x+2*h) + 4*f(x+h) - 3*f(x))/(2*h)


print('df(%.2f)/dx = %.4f' %(x,df))
```

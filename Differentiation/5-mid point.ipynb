```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*exp(x)

h=0.01
x=2
df  = (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x+2*h))/(12*h)

print('df(%.2f)/dx = %.4f' %(x,df))
```

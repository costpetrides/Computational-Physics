```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
  return x*np.exp(x)

h=0.01
x=2

ddf = (f(x-h) - 2*f(x) + f(x+h))/(h**2)

print('df(%.2f)/dx = %.4f' %(x,ddf))
```

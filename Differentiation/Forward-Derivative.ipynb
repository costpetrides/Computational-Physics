**Example 1:** Find the first derivative of $f(x)= x^3 + \frac{5e^{x}}{x}$ at x=2

```python
import numpy as np
import matplotlib.pyplot as plt

h=0.04

def f(x):
  return x**3+5*np.exp(x)/x

def df(x):
  return (f(x+h)-f(x))/h

def ddf(x):
  return (f(x-h)-2*f(x)+f(x+h))**2/h**2

Error=(-h/2)*ddf(2)
print('df(2)/dx= %.3f +/- %.3f' %(df(2),Error))
```

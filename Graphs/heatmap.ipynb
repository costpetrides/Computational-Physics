Plotting a 2D heatmap with Matplotlib

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd

array = [[23, 8, 5, 13, 11, 13, 7, 36, 28, 20, 23, 10],    # G: years and S: months
         [9, 14, 17, 15, 5, 12, 16, 30, 21, 8, 8, 6],
         [9, 11, 24, 40, 19, 29, 42, 36, 15, 7, 14, 10],
         [8, 13, 19, 19, 12, 15, 41, 33, 30, 26, 14, 11]]

month=['Jan', 'Feb', 'Mar*', 'Apr*', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec']      
year=['2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('O$_{3}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'μg/$m^{3}$'}, cmap='Blues')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```

The increase in $O_{3}$ concentrations during the  lockdown (Covid-19) could be attributed to the increase in hydrocarbons and nitrogen oxidesratios since atmospheric chemistry in Athens  is under VOC-controlled conditions.

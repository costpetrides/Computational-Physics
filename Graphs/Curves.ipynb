**Plot 2D Curve**

$\vec{r}(t) = \frac{1}{2}cos(t) \hat{i} + sin(2t)\hat{j}  \quad    0 \leq t \leq 10     $

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 200)        # Πεδίο ορισμού:  0 <= t <=10  200 στοιχεία
      
x = 1/2* np.cos(t)                 # x(t)                       
y = np.sin(2*t)                    #y(t)    
                                     
plt.plot(x, y)                     # Κατασκευή Γραφικής  
plt.show()
```

**Plot  3D Curve**

$\vec{r}(t)= cos(t) \hat{i} + sin(t) \hat{j} + t \hat{k} \quad  0 \leq t \leq 10  $

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 6))      # Καθορισμός διαστάσεων εικόνας     
ax = plt.axes(projection="3d")         # 3D Γραφικά

t = np.linspace(0, 10, 1000)           #  Πεδίο ορισμού:  0 <= t <=10  100 στοιχεία 

x = np.cos(t)                          #  x = cos(t) 
y = np.sin(t)                          #  y = sin(t)
z = t                                  #  z = t      

ax.plot3D(x, y, z, 'gray')             # Κατασκευή Γραφικής
plt.show()
```

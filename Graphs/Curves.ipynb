**Plot 2D Curve**

$\vec{r}(t) = \frac{1}{2}cos(t) \hat{i} + sin(2t)\hat{j}  \quad    0 \leq t \leq 10     $

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 200)        # Πεδίο ορισμού:  0 <= t <=10  200 στοιχεία

x = 1/2* np.cos(t)                 # x(t)
y = np.sin(2*t)                    #y(t)

plt.plot(x, y)                     # Κατασκευή Γραφικής
plt.title('Title')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xlim(-0.6,0.6)                 # Όρια x άξονα
plt.ylim(-1.1,1.1)                 # Όρια y άξονα
plt.show()
```

**Plot  3D Curve**

$\vec{r}(t)= cos(t) \hat{i} + sin(t) \hat{j} + t \hat{k} \quad  0 \leq t \leq 10  $

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(13, 7))      # Καθορισμός διαστάσεων εικόνας
ax = plt.axes(projection="3d")         # 3D Γραφικά

t = np.linspace(0, 10, 1000)           #  Πεδίο ορισμού:  0 <= t <=10  100 στοιχεία

x = np.cos(t)                          #  x = cos(t)
y = np.sin(t)                          #  y = sin(t)
z = t                                  #  z = t

ax.plot3D(x, y, z, 'blue')             # Κατασκευή Γραφικής
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-1.5,1.5)                 # Όρια x άξονα
ax.set_ylim(-1.5,1.5)                 # Όρια y άξονα
ax.set_zlim(0,10)
plt.show()
```

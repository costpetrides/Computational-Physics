Vector Field - 2D

$ \vec{F}= -\frac{y}{\sqrt{x^{2} + y^{2}}} \hat{i} + \frac{x}{\sqrt{x^{2} + y^ {2}}} \hat{j} $ 

$\vec{F} = u(x,y) \hat{i} + v(x,y) \hat{j}$

```python
import numpy as np                        # Βιβλιοθήκες 
import matplotlib.pyplot as plt

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))    # Ορίζω ορθογώνιο πλέγμα  
                                                                    #[-10,10] x [-10,10] 
                                                                    #με 20 στοιχεία το καθένα     

u = -y/np.sqrt(x**2 + y**2)        # Ορίζω το u(x,y)  
v =  x/np.sqrt(x**2 + y**2)        # Ορίζω το v(x,y)  

plt.quiver(x,y,u,v)                # Κατασκευή γραφικής 
plt.show()
```

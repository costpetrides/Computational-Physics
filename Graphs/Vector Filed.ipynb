Vector Field - 2D

$ \vec{F}= -\frac{y}{\sqrt{x^{2} + y^{2}}} \hat{i} + \frac{x}{\sqrt{x^{2} + y^ {2}}} \hat{j} $ 

$\vec{F} = u(x,y) \hat{i} + v(x,y) \hat{j}$

```python
import numpy as np                        #  
import matplotlib.pyplot as plt

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))    #     
                                                                    #[-10,10] x [-10,10] 
                                                                    # 20        

u = -y/np.sqrt(x**2 + y**2)        #   u(x,y)  
v =  x/np.sqrt(x**2 + y**2)        #   v(x,y)  

plt.quiver(x,y,u,v)                #   
plt.show()
```

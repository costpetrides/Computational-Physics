```python
import numpy as np
import matplotlib.pyplot as plt
import math



count=0

def u():
  return np.random.rand()

N=2000
for i in range(N):
  x=u()
  y=u()
  if x<=0.5 and y<=0.5:
    plt.scatter(x, y,c='green')
    count+=1
  else:
    plt.scatter(x, y,c='blue')

plt.show()
p=count/N
print('The percentage is : %.2f' %(p*100), '%')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math



count1=0

def u():
  return np.random.rand()


for j in range(N):
  x1=u()
  y1=u()
  if x1**2 + y1**2 <=1:
    plt.scatter(x1, y1,c='green')
    count1+=1
  else:
    plt.scatter(x1, y1,c='blue')

plt.show()
p1=count1/N
print('The percentage is : %.2f' %(p1*100), '%')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

cc=0

def uu(a,b):
  return (b-a)*np.random.rand() +a


for i in range(N):
  xx=uu(-1,1)
  yy=uu(0,1)
  if yy**2 + xx**2 <=1:
    plt.scatter(xx, yy,c='green')
    cc+=1
  else:
    plt.scatter(xx, yy,c='red')

plt.show()
p2=cc/N
print('The percentage is : %.2f' %(p2*100), '%')
```

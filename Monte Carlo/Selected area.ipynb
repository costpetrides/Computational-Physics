**Find the probability that a coin lands on the green area?**



**a)**

IMG_0839.jpg

```python
import numpy as np
import matplotlib.pyplot as plt
import math



count=0

def u():
  return np.random.rand()

N=2000
for i in range(N):
  x=u()
  y=u()
  if x<=0.5 and y<=0.5:
    plt.scatter(x, y,c='green')
    count+=1
  else:
    plt.scatter(x, y,c='red')

plt.show()
p=count/N
print('The percentage is : %.2f' %(p*100), '%')
```

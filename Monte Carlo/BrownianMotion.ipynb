```python
import numpy as np
import matplotlib.pyplot as plt

def di():
  return np.random.randint(1, 5)


xw=[]      # Λίστα τελικής θέσης κάθε σωμάτιου  x-axis
yw=[]      # Λίστα τελικής θέσης κάθε σωμάτιου  y-axis
sr=0
sr2=0
for i in range(10000):      # 10000 σωμάτια
  x=[0]                    # Λίστα θέσης σωματιδίου σε κάθε βήμα
  y=[0]
  for j in range(100):    # Bήματα
    k=di()                # " Απόφαση κίνησης σωματιδίου"
    if k==1:
      x+= [1 + x[-1]]
      y+= [y[-1]]
    elif k==2:
      x+= [-1 + x[-1]]
      y+= [y[-1]]
    elif k==3:
      x+= [x[-1]]
      y+= [1 + y[-1]]
    else:
      x+= [x[-1]]
      y+= [-1 + y[-1]]

  xw +=[x[-1]]
  yw +=[y[-1]]



xwa=np.asarray(xw)
ywa=np.asarray(yw)
d = sum(np.sqrt(xwa**2 + ywa**2))/10000.     # Mέσος όρος  απόστασης
d2= sum(xwa**2 + ywa**2)/10000               # Μέσο όρο του τετραγώνου της απόστασης


plt.figure(figsize=(12,5))
plt.hist(xw, bins=100, alpha=0.7, label='x-axis')
plt.hist(yw, bins=100, alpha=0.7, label='y-axis')
plt.legend(loc='upper right')
plt.title('Final position of particle')
plt.show()

plt.figure(figsize=(12,5))
plt.scatter(xw,yw)
plt.title('Final gas state')
plt.show()

plt.figure(figsize=(12,5))
plt.plot(x,y,label='One particle motion (i=10000) για t=100')
plt.xlabel('Position x-axis')
plt.ylabel('Position y-axis')
plt.legend()
plt.show()

print('2D Brownian motion:')

print('Average distance of the particles after t=100 sec = %.2f' %(d))
```

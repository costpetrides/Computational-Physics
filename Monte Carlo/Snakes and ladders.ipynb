```python
import numpy as np
import matplotlib.pyplot as plt
import math

def u():
  return int(6*np.random.rand()+1)

def game():
  pos=1
  hit=0
  while pos <=36:
    hit+=1
    pos+=u()
    if pos==3:
      pos=16
    if pos==12:
      pos=2
    if pos==15:
      pos=25
    if pos==21:
      pos=32
    if pos==30:
      pos=4
    if pos==35:
      pos=22
  return hit



x=[game() for i in range(1000)]
plt.hist(x,bins=range(1,60),ec='black')
```

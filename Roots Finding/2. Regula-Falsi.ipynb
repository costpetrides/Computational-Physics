```python
# y=x^3-x-1 , [1,2]
import numpy as np
from  math import *
import matplotlib.pyplot as plt

def f(x):
  return x**3-x-1

a=1
b=2

data=[]

for i in range(100):
  c = (f(b)*a-f(a)*b)/(f(b)-f(a))
  if (f(a))*f(c) <0 :
    b = c
  else:
   a = c
  if abs(f(c)) < 1.e-6:
    break
  print('N=',i, 'Sol:',c)
  data+=[c]

x=np.linspace(1,i,i)
plt.plot(x,data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

def f(x):
  return x**3-x-1
x=[]
a=1
b=2
print('Bisection')
for i in range(100):
  c=(a+b)/2
  if f(a)*f(c)<0:
    b=c
  else:
    a=c
  x+=[c]
  print('N[%.f]= %.5f' %(i+1,c))
  if abs(f(c))<1.e-5:
    break

plt.plot(x,label='Bisection , N=%.f' %(i+1))

print('')
print('')

print('Regula falsi')
x1=[]
a1=1
b1=2
for i in range(100):
  c1=(a1*f(b1)-b1*f(a1))/(f(b1)-f(a1))
  if f(a1)*f(c1)<0:
    b1=c1
  else:
    a1=c1
  x1+=[c1]
  print('N[%.f]= %.5f' %(i+1,c1))
  if abs(f(c1))<1.e-5:
    break


plt.plot(x1,label='Regula falsi , N=%.f' %(i+1))
plt.legend()
plt.show()
```

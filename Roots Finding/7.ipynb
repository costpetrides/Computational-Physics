```python
import numpy as np
import matplotlib.pyplot as plt


x=3
new=[]

def f(x):
  return x**4 - 4*x**3 +4*x**2 + x - 4

def df(x):
  return 4*x**3-12*x**2+8*x+1

for i in range(1000):
  x=x-(f(x)/df(x))
  new+=[x]
  if abs(f(x))<1.e-6:
    break

N=len(new)

print('Newton Raphson: x= %.6f'  %(new[-1]))
print('N= %.f' %(N))
plt.plot(new)
plt.axhline(y = 2.492573, c='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ορισμός της συνάρτησης f και της παραγώγου της
def f(x):
    return x**4 - 4*x**3 + 4*x**2 + x - 4

def df(x):
    return 4*x**3 - 12*x**2 + 8*x + 1

# Διάστημα αναζήτησης
a, b = 2, 3
x0 = None

# === Βήμα 2 === Επιλογή κατάλληλης αρχικής τιμής
if f(a) * df(a) > 0:
    x0 = a
elif f(b) * df(b) > 0:
    x0 = b
else:
    raise ValueError("Δεν υπάρχει κατάλληλη αρχική τιμή που να ικανοποιεί f(x)·f'(x) > 0")

print(f"Αρχική τιμή x₀ επιλέχθηκε το x₀ = {x0}")

# === Βήμα 3 === Επαναληπτική διαδικασία
x = x0
newton_vals = []

for i in range(1000):
    x_new = x - f(x) / df(x)
    newton_vals.append(x_new)
    if abs(f(x_new)) < 1e-6:
        break
    x = x_new  # ενημέρωση x

# Αποτελέσματα
print("Newton-Raphson: x = %.6f" % newton_vals[-1])
print("Αριθμός επαναλήψεων:", len(newton_vals))

# Γράφημα
plt.plot(newton_vals, marker='o', label='Προσεγγίσεις ρίζας')
plt.axhline(y=newton_vals[-1], color='red', linestyle='--', label='Τελική τιμή')
plt.xlabel('Επανάληψη')
plt.ylabel('Προσεγγιστική ρίζα x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4 - 4*x**3 + 4*x**2 + x - 4

def df(x):
    return 4*x**3 - 12*x**2 + 8*x + 1

a, b = 2, 3  # Διάστημα αναζήτησης ρίζας

# Έλεγχος για την επιλογή αρχικής τιμής x0
if f(a) * df(a) > 0:
    x = a
elif f(b) * df(b) > 0:
    x = b
else:
    raise ValueError("Δεν βρέθηκε κατάλληλη αρχική τιμή που να ικανοποιεί f(x)*f'(x) > 0")

new = []

for i in range(1000):
    x = x - f(x)/df(x)
    new.append(x)
    if abs(f(x)) < 1.e-6:
        break

N = len(new)

print('Newton-Raphson: x= %.6f' % new[-1])
print('Αριθμός επαναλήψεων: %d' % N)

plt.plot(new, marker='o')
plt.axhline(y=new[-1], color='red', linestyle='--', label='Τελική τιμή')
plt.xlabel('Επανάληψη')
plt.ylabel('Προσεγγιστική ρίζα x')
plt.legend()
plt.grid(True)
plt.show()
```

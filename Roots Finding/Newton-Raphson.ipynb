```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**4 -4*x**3 +4*x**2 + x -4

def df(x):
  return 4*x**3 -12*x**2 +8*x + 1

x=3
root=[]
for i in range(1000):
  x=x-(f(x)/df(x))
  root+=[x]
  print('N=',i,'Sol=',x)
  if abs(f(x))<=1.e-6:
    break

plt.plot(root)
plt.show()
```

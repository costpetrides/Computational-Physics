```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 - x-1

def g(x):
  return (x+1)**(1/3)

x=1
root=[]
for i in range(100):
  x=g(x)
  root+=[x]
  print('N=',i,'Sol=',x)
  if abs(f(x))<=1.e-6:
    break

plt.plot(root)
plt.show()
```

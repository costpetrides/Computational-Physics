**$y=x^3 -x-1$ initial guess $x_{o}=2$**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 -x-1

def df(x):
  return 3*x**2-1

x=2
root=[]
for i in range(1000):
  x=x-(f(x)/df(x))
  root+=[x]
  print('N=',i,'Sol=',x)
  if abs(f(x))<=1.e-6:
    break

plt.plot(root)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

#Bisection

def f(x):
  return x**3-x-1
x=[]
a=1
b=2
print('Bisection')
for i in range(100):
  c=(a+b)/2
  if f(a)*f(c)<0:
    b=c
  else:
    a=c
  x+=[c]
  print('N[%.f]= %.5f' %(i+1,c))
  if abs(f(c))<1.e-5:
    break

plt.plot(x,label='Bisection , N=%.f' %(i+1))

print('')
print('')

#Regula Falsi
print('Regula falsi')
x1=[]
a1=1
b1=2
i=0
for i in range(100):
  c1=(a1*f(b1)-b1*f(a1))/(f(b1)-f(a1))
  if f(a1)*f(c1)<0:
    b1=c1
  else:
    a1=c1
  x1+=[c1]
  print('N[%.f]= %.5f' %(i+1,c1))
  if abs(f(c1))<1.e-5:
    break

plt.plot(x1,label='Regula falsi , N=%.f' %(i+1))

print('')
print('')


# Newton - Raphson
print('Newton - Raphson')
def df(x):
  return 3*x**2 - 1

p=[]
x=1
i=0
for i in range(100):
  x=x-f(x)/df(x)
  p+=[x]
  print('N[%.f]= %.6f' %(i+1,x))
  if abs(f(x))<1.e-6:
    break


plt.plot(p,label='Newton N=%.f' %(i+1))

plt.legend()


print('')
print('')

#Fixed Point

print('Fixed Point')

def g(x):
  return (1+x)**(1/3)

i=0
x3=[]
x=1


for i in range(100):
  x=g(x)
  x3+=[x]
  print('N[%.f]= %.5f' %(i+1,x))
  if abs(f(x))<1.e-5:
    break

plt.plot(x3,label='Fixed Point, N=%.f' %(i+1))
plt.legend()


print('')
print('')

#Secant 

print('Secant')


a=1
b=2
i=0
x4=[]
for i in range(100):
  c=(a*f(b)-b*f(a))/(f(b)-f(a))
  x4+=[c]
  print('N[%.f] = %.5f ' %(i+1,c))
  if abs(f(c))<1.e-6:
    break
  else:
    a=b
    b=c

plt.plot(x4,label='Secant,  N=%.f' %(i+1))
plt.legend()

plt.show()
```

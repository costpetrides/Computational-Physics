**Secant Method**

secant.PNG


```python
# y=x^3-x-1   
import numpy as np
from math import *

def f(x):
  return x**3-x-1

a=1
b=2
for i in range(100):
  c=(a*f(b)-b*f(a))/(f(b)-f(a))
  if abs(f(c))<1.e-6:
    print('N=',i,'The root is:',c )
    break
  else:
    a=b
    b=c
```

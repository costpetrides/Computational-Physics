```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

def f(x):
    return x**4 -4*x**3 +4*x**2 +x -4

def df(x):
    return 4*x**3 -12*x**2 +8*x +1

a=2
b=3
Root=[]
for i in range(100):
  c = a - (a - b) * f(a)/ (f(a) - f(b))
  print('N=%.f : Root=%.3f' %(i,c))
  Root+=[c]
  if abs(f(c))<=1.e-6:
    break
  a = b
  b = c 

plt.plot(Root)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 + 2*x**2 - 3*x - 1

bis = []
a = 1
b = 2

for i in range(1000):
  c = (a + b) / 2
  if f(a) * f(c) < 0:
    b = c
  elif f(c) * f(b) < 0:
    a = c
  bis.append(c)
  if abs(f(c)) < 1.e-6:
    break

N = len(bis)

print('Bisection: x= %.6f' % bis[-1])
print('N= %d' % N)

plt.plot(bis, marker='o', label='Προσεγγίσεις ρίζας')
plt.axhline(y=1.1875, color='red', linestyle='--', label='Τελική τιμή c3')
plt.xlabel('Επανάληψη')
plt.ylabel('Προσεγγιστική ρίζα c')
plt.legend()
plt.grid(True)
plt.show()

```

# Υπολογιστικά οι επαναλήψεις

```python
def f(x):
    return x**3 + 2*x**2 - 3*x - 1

a = 1
b = 2
tol = 1e-6
max_iter = 4  # τέσσερις επαναλήψεις

for i in range(max_iter):
    c = (a + b) / 2
    fc = f(c)
    print(f"Επανάληψη {i+1}: a={a:.6f}, b={b:.6f}, c={c:.6f}, f(c)={fc:.6f}")

    if abs(fc) < tol:
        print(f"Σύγκλιση με |f(c)| < {tol}")
        break

    if f(a) * fc < 0:
        b = c
    else:
        a = c
```

```python
from scipy.optimize import brentq

def f(x):
    return x**3 + 2*x**2 - 3*x -1

root = brentq(f, 1, 2)
print(f"Ακριβής ρίζα: {root:.8f}")
```

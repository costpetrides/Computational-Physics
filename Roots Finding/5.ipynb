```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 + 2*x**2 - 3*x - 1


bis=[]
a=1
b=2
c=0
for i in range(1000):
  c=(a+b)/2
  if f(a)*f(c)<0:
    b=c
  if f(c)*f(b)<0:
    a=c
  bis+=[c]
  if abs(f(c))<1.e-6:
    break

N=len(bis)

print('Bisection: x= %.6f'  %(bis[-1]))
print('N= %.f' %(N))
plt.plot(bis)
plt.axhline(y = 1.324718, c='red')
plt.show()
```

```python
import numpy as np
from  math import *

def f(x):
  return x**3-x-1

a=1
b=2
for i in range(100):
  c=(a+b)/2
  if f(a)*f(c)<0:
    b=c
  else:
    a=c
  if abs(f(c))<1.e-6:
   break
  print('N=',i,'Sol=',c)
```

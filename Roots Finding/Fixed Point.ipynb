```python
# f(x)=x**2-x-2     sol: 2 and -1. The g(x)=(2+x)^0.5 ---> 2
import numpy as np
from math import *

def f(x):
  return x**2-x-2

def g(x):
  return (2+x)**0.5

x=1
for i in range(100):
  x=g(x)
  if abs(f(x)) < 1.e-6: 
    break
  print('N=',i,'x=',x)
```

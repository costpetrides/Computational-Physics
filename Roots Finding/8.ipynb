```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 2*x**2 - 3*x - 1

# Αρχικές τιμές από το διάστημα (1,2)
x0 = 1
x1 = 2
secant = []

for i in range(1000):
    fx0 = f(x0)
    fx1 = f(x1)

    if fx1 - fx0 == 0:
        raise ZeroDivisionError("Διαίρεση με μηδέν στη μέθοδο της τέμνουσας.")

    x2 = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
    secant.append(x2)

    if abs(f(x2)) < 1e-6:
        break

    # Ενημέρωση τιμών
    x0, x1 = x1, x2

# Τελικό αποτέλεσμα
print("Secant method: x = %.6f" % x2)
print("Αριθμός επαναλήψεων: %d" % len(secant))

# Γράφημα
plt.plot(secant, marker='o', label='Προσεγγίσεις ρίζας')
plt.axhline(y=secant[-1], color='red', linestyle='--', label='Τελική τιμή')
plt.xlabel('Επανάληψη')
plt.ylabel('Προσεγγιστική ρίζα x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
  return x**3 +2*x**2 - 3*x - 1

x2=1.3
x1=1
x0=2
sec=[]

for i in range(100):
  x2 = x1 - f(x1)*(x1-x0)/( f(x1) - f(x0))
  x0=x1
  x1=x2
  sec+=[x2]
  if abs(f(x2))<1.e-7:
    break
N=len(sec)

print('Secant: x= %.6f'  %(sec[-1]))
print('N= %.f' %(N))
plt.plot(sec)
plt.axhline(y = 1.1986912435, c='red')
plt.show()
```

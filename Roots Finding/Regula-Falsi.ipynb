$y=x^3 -x-1$ ,[1,2]

```python
import matplotlib.pyplot as plt
import numpy as np
from  math import *

def f(x):
  return x**3-x-1

root=[]

a=1
b=2
for i in range(100):
  c=(a*f(b) -b*f(a))/(f(b)-f(a))
  if f(a)*f(c)<0:
    b=c
  else:
    a=c
  root+=[c]
  print('N=',i,'Sol=',c)
  if abs(f(c))<1.e-6:
   break
  
plt.plot(root)
plt.show()
```

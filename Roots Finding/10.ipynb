```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return  x**3 + 2*x - 1

def g(x):
  return (1 - x**3) * 0.5

x = 0  # αρχική τιμή
fix = []

for i in range(100):
  x = g(x)
  fix.append(x)
  if abs(f(x)) < 1.e-6:
    break

N = len(fix)

print('Fixed Point: x= %.6f' % fix[-1])
print('Αριθμός επαναλήψεων: %d' % N)

plt.plot(fix, marker='o')
plt.axhline(y=fix[-1], color='red', linestyle='--')
plt.xlabel('Επανάληψη')
plt.ylabel('Προσεγγιστική ρίζα x')
plt.title('Μέθοδος Σταθερού Σημείου')
plt.grid(True)
plt.show()
```

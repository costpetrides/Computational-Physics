```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return  x**3 +2*x-1

x=0
fix=[]

def g(x):
  return (1-x**3)*(1/2)

for i in range(100):
  x=g(x)
  fix+=[x]
  if abs(f(x))<1.e-6:
    break

N=len(fix)

print('Newton Raphson: x= %.6f'  %(fix[-1]))
print('N= %.f' %(N))
plt.plot(fix)
plt.axhline(y = 0.453398, c='red')
plt.show()
```

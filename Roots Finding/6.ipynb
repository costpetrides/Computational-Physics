```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 2*x**2 - 3*x - 1

reg = []
a, b = 1, 2

for i in range(1000):
    c = (a*f(b) - b*f(a)) / (f(b) - f(a))
    if f(a)*f(c) < 0:
        b = c
    elif f(c)*f(b) < 0:
        a = c
    reg.append(c)
    if abs(f(c)) < 1.e-6:
        break

N = len(reg)

print(f'Regula Falsi: x= {reg[-1]:.6f}')
print(f'N= {N}')

plt.plot(reg, marker='o')
plt.axhline(y=1.19869, color='red', linestyle='--', label='Τελική τιμή')
plt.xlabel('Επανάληψη')
plt.ylabel('Προσεγγιστική ρίζα c')
plt.legend()
plt.grid(True)
plt.show()

```

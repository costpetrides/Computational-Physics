```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 + 2*x**2 - 3*x - 1

reg=[]
a=1
b=2
c=0
for i in range(1000):
  c=(a*f(b)-b*f(a))/(f(b)-f(a))
  if f(a)*f(c)<0:
    b=c
  if f(c)*f(b)<0:
    a=c
  reg+=[c]
  if abs(f(c))<1.e-6:
    break

N=len(reg)

print('Regula Falsi: x= %.6f'  %(reg[-1]))
print('N= %.f' %(N))
plt.plot(reg)
plt.axhline(y = 1.19869, c='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

v=[]

file=open("data.txt","r")
N=0

for x in file:
  v.append(float(x))
  N+=1

m=np.sum(v)/N
sd=np.sum((v-m)**2)/N

print(m)
print(sd)
```

```python
v=[]

file=open("data1.txt","r")
N=0

for x in file:
  v.append(float(x))
  N+=1

m1=np.sum(v)/N
sd1=np.sum((v-m1)**2)/N

print(m1)
print(sd1)
```

```python
v=[]

file=open("data2.txt","r")
N=0

for x in file:
  v.append(float(x))
  N+=1

m2=np.sum(v)/N
sd2=np.sum((v-m2)**2)/N

print(m2)
print(sd2)
```

```python
mean=[m,m1,m2]
names=['Data','Data1','Data2']

fig = plt.figure(figsize=(4,2), dpi=200)
left, bottom, width, height = 0.1, 0.2, 0.8, 0.6
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.5
ticks = np.arange(len(names))    

ax.bar(ticks, mean, width,label=' <x> from different files')


ax.set_ylabel('mean value for x')
ax.set_title('Mean value from different files')
ax.set_xticks(ticks + width/45)
ax.set_yticks((0,50,100,150,200,250))
ax.set_xticklabels(names)

ax.legend(loc='best')
plt.show()
```

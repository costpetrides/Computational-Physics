# Linear Gongruential Generator
---



1.   Give an initial value: $x_{i}$      $⇒$ The first number is not a "random' one

2.   Generator →   $x_{i+1}= (a * x_{i} + b)\% m$ 

      where a, b and m are constants 



```python
import numpy as np
import matplotlib.pyplot as plt
import time


def lcg(x,a,b,m,num_rand):
  gen=[]
  for i in range(num_rand):
    x=(a*x+b)%m
    gen.append(x)
  return np.array(gen)

tim=time.perf_counter()
numbers=lcg(tim,15,6,7,100)
print(numbers)


plt.plot(np.linspace(0,100,100), numbers)
```

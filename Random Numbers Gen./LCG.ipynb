** It can fall into a loop **

```python
import numpy as np
import matplotlib.pyplot as plt
import time


def lcg(x,a,b,m,num_rand):
  gen=[]
  for i in range(num_rand):
    x=(a*x+b)%m
    gen.append(x)
  return np.array(gen)

tim=time.perf_counter()
numbers=lcg(tim,15,6,7,100)
print(numbers)


plt.plot(np.linspace(0,100,100), numbers)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u(a, b):
    """Παράγει έναν τυχαίο αριθμό από ομοιόμορφη κατανομή [a, b]."""
    return np.random.uniform(a, b)

c = np.exp(5) - 1

def f(x):
    """Συνάρτηση κατανομής f(x) = e^x / c."""
    return np.exp(x) / c

fmax = f(5)

numbers = []

for _ in range(10000):
    x = u(0, 5)
    y = u(0, fmax)
    if y <= f(x):
        numbers.append(x)

plt.hist(numbers, bins=20, edgecolor='black') # Επέλεξα bins=20 για ομαλή και λεπτομερή απεικόνιση της κατανομής
plt.title('Hit or Miss Sampling για $f(x) = e^x / c$')
plt.xlabel('x')
plt.ylabel('Συχνότητα')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# Προσοχή !!

Ο αριθμός των bins στο ιστόγραμμα δεν επηρεάζει την ίδια τη φυσική ή τα μαθηματικά της κατανομής των δεδομένων. Αντιθέτως, επηρεάζει μόνο τον τρόπο με τον οποίο απεικονίζονται οπτικά τα δεδομένα, δηλαδή το πόσο λεπτομερές ή χονδροειδές εμφανίζεται το γράφημα. Επίσης, το πώς κατανέμονται οι αριθμοί στα bins (δηλαδή σε ποιο «δοχείο» τοποθετείται κάθε τιμή) καθορίζεται από τα όρια των bins που ορίζει ο χρήστης ή η προεπιλεγμένη μέθοδος της βιβλιοθήκης. Η σωστή επιλογή αριθμού και ορίων bins βοηθά στην καλύτερη οπτική κατανόηση της κατανομής, αλλά δεν αλλάζει τα υπολογισμένα στατιστικά ή τις ιδιότητες της κατανομής.


```python
import numpy as np
import matplotlib.pyplot as plt

def u(a, b):
    """Παράγει έναν τυχαίο αριθμό από ομοιόμορφη κατανομή [a, b]."""
    return np.random.uniform(a, b)

c = np.exp(5) - 1

def f(x):
    """Συνάρτηση κατανομής f(x) = e^x / c."""
    return np.exp(x) / c

fmax = f(5)

numbers = []

for _ in range(10000):
    x = u(0, 5)
    y = u(0, fmax)
    if y <= f(x):
        numbers.append(x)

plt.hist(numbers, bins=5, edgecolor='black') # Εδώ επέλεξα  bins=5
plt.title('Hit or Miss Sampling για $f(x) = e^x / c$')
plt.xlabel('x')
plt.ylabel('Συχνότητα')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

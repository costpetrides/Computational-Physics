```python
import numpy as np
import matplotlib.pyplot as plt
```

##Exercise 1:
### i.

```python
a=1.0362
b=3.1753
m=4
x=1.673
numbers=[]

for i in range(100):
  x=(a*x+b)%m
  numbers+=[x]

plt.hist(numbers,bins=np.linspace(0,4,8),ec='black')
plt.show()
```

### ii.

```python
def uniform(a,b):
  return (b-a)*np.random.rand()+a

numbers=[uniform(0,4) for i in range(100)]

plt.hist(numbers,bins=np.linspace(0,4,8),ec='black')
plt.show()
```

##Exercise 2:
### i.

```python
def u():
  return np.random.rand()

def Finv(u):
  return np.sqrt(u)

x=[Finv(u()) for i in range(1000)]
plt.hist(x,bins=np.linspace(0,1,10),ec='black')
plt.show()
```

### ii.

```python
def u():
  return np.random.rand()

def Finv(u):
  return np.sqrt(3*u+1)

x=[Finv(u()) for i in range(1000)]
plt.hist(x,bins=np.linspace(1,2,10),ec='black')
plt.show()
```

### iii.

```python
def u():
  return np.random.rand()

def Finv(u):
  c=3/26
  return ((3*u+c)/c)**(1/3)

x=[Finv(u()) for i in range(1000)]
plt.hist(x,bins=np.linspace(1,3,20),ec='black')
plt.show()
```

##Exercise 3:


```python
def u():
  return np.random.rand()

xhit=[]
yhit=[]
for i in range(1000):
  x=(2+2)*u()-2
  y=(2+2)*u()-2
  if x**2 +y**2<=4:
    xhit+=[x]
    yhit+=[y]

plt.scatter(xhit,yhit)
plt.show()
```

##Exercise 4:


```python
import numpy as np
import matplotlib.pyplot as plt

def u():
  return np.random.rand()

def f(x):
  c=1/np.log(21)
  return c/x

a=0.5
b=10.5
fmax=f(a)

xhit=[]
yhit=[]
for i in range(10000):
  x=(b-a)*u()+a
  y=fmax*u()
  if y<=f(x):
    xhit+=[x]
    yhit+=[y]


plt.hist(xhit,bins=np.linspace(a,b,10),ec='black')
plt.show()
plt.scatter(xhit,yhit,ec='green')
plt.show()
```

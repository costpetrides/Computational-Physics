```python
import numpy as np
import matplotlib.pyplot as plt

# Σταθερά κανονικοποίησης
c = np.exp(5) - 1

def u():
    """Παράγει έναν τυχαίο αριθμό από ομοιόμορφη κατανομή [0,1)."""
    return np.random.rand()

def Finv(u_val):
    """Αντίστροφη αθροιστική κατανομή F^{-1}(u)."""
    return np.log(1 + u_val * c)

# Δημιουργία 1000 τυχαίων δειγμάτων από την κατανομή e^x στο [0,5]
samples = [Finv(u()) for _ in range(1000)]

# Σχεδίαση ιστογράμματος
plt.hist(samples, bins=30, edgecolor='black')
plt.title("Δειγματοληψία Αντίστροφου Μετασχηματισμού")
plt.xlabel("Τιμές")
plt.ylabel("Συχνότητα")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

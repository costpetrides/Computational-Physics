```python
import numpy as np
import matplotlib.pyplot as plt
x=[]

c=np.exp(5)-1

def u():
  return np.random.rand()   #Function that gives one radnom  number (0,1)

def Finv(x):
  return np.log(1 +x*c)

x=[Finv(u()) for i in range(1000)]         # Tips!!
plt.hist(x, ec='black')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data=[]

def u(a,b):
  return (b-a)*np.random.rand() + a

c=np.exp(5)-1

def f(x):
  return np.exp(x)/c

fmax=f(5)
for i in range(10000):
  x=u(0,5)
  y=u(0,fmax)
  if y<=f(x):
    data+=[x]

plt.hist(data, ec='black')
plt.show()
```

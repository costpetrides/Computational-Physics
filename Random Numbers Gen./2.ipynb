# i.

```python
import matplotlib.pyplot as plt

a = 1103515245
c = 12345
m = 231   #   i. 
x = 2

N = 100
numbers = []

for _ in range(N):
    x = (a * x + c) % m
    numbers.append(x)

plt.plot(numbers, marker='o')
plt.title('LCG -  i.')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

# ii.

```python
import matplotlib.pyplot as plt

a = 1103515245
c = 12345
m = 2**31
x = 2


N = 100
numbers = []

for _ in range(N):
    x = (a * x + c) % m
    numbers.append(x)

plt.plot(numbers, marker='o')
plt.title('LCG -  ii.')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

#  ,               [0,1),         m.

```python
normalized_numbers = [num / m for num in numbers]

plt.plot(normalized_numbers, marker='o')
plt.title('LCG (normalized) -  ii.')
plt.xlabel('')
plt.ylabel(' (0  1)')
plt.grid(True)
plt.show()
```

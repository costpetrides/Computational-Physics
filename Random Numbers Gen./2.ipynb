# i.

```python
import matplotlib.pyplot as plt

a = 1103515245
c = 12345
m = 231   # Για το i. παράδειγμα
x = 2

N = 100
numbers = []

for _ in range(N):
    x = (a * x + c) % m
    numbers.append(x)

plt.plot(numbers, marker='o')
plt.title('LCG - Παράδειγμα i.')
plt.xlabel('Βήμα')
plt.ylabel('Τιμή')
plt.grid(True)
plt.show()
```

# ii.

```python
import matplotlib.pyplot as plt

a = 1103515245
c = 12345
m = 2**31
x = 2


N = 100
numbers = []

for _ in range(N):
    x = (a * x + c) % m
    numbers.append(x)

plt.plot(numbers, marker='o')
plt.title('LCG - Παράδειγμα ii.')
plt.xlabel('Βήμα')
plt.ylabel('Τιμή')
plt.grid(True)
plt.show()
```

# Αν θέλουμε, μπορούμε να κάνουμε και κανονικοποίηση ώστε οι τυχαίοι αριθμοί να είναι μέσα στο διάστημα [0,1), αρκεί απλά να διαιρέσουμε κάθε αριθμό με το m.

```python
normalized_numbers = [num / m for num in numbers]

plt.plot(normalized_numbers, marker='o')
plt.title('LCG (normalized) - Παράδειγμα ii.')
plt.xlabel('Βήμα')
plt.ylabel('Τιμή (0 έως 1)')
plt.grid(True)
plt.show()
```

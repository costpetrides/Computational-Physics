```python
import matplotlib.pyplot as plt

a =1103515245
c = 12345
m =2**31
x=2

N=100
numbers=[]
for i in range(N):
  x=(a*x+c)%m
  numbers+=[x]

plt.plot(numbers)
plt.show()
```

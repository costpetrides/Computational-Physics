```python
import numpy as np
import matplotlib.pyplot as plt

def u():
  return np.random.rand()

def f(x):
  return np.exp(x)/(np.exp(5)-1)

hit=0
xhit=[]
yhit=[]
xmiss=[]
ymiss=[]

fmax=f(5)
for i in range(10000):
  x=5*u()
  y=fmax*u()
  if y<f(x):
    hit+=1
    xhit+=[x]
    yhit+=[y]
  else:
    xmiss+=[x]
    ymiss+=[y]

bin=np.linspace(0,5,50)
plt.hist(xhit,bins=bin, ec='black')
plt.show()

print('Integral value =%.3f' %(hit))  # Also we can find the value of integral !!
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u():
  return np.random.rand()

def f(x):
  return np.exp(x)/(np.exp(5)-1)

hit=0
xhit=[]
yhit=[]
xmiss=[]
ymiss=[]

fmax=f(5)
for i in range(1000):
  x=5*u()
  y=fmax*u()
  if y<f(x):
    hit+=1
    xhit+=[x]
    yhit+=[y]
  else:
    xmiss+=[x]
    ymiss+=[y]

plt.scatter(xhit,yhit,c='green')
plt.scatter(xmiss,ymiss,c='red')
plt.show()

print('I=%.3f' %(hit))    # or we can use the len(hit)
```

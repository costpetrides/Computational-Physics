```python
import numpy as np
import matplotlib.pyplot as plt

def u():
  return np.random.rand()

def Finv(u):
  return np.log(1+u*(np.exp(5) -1))

x=[Finv(u()) for i in range(1000)]

bin=np.linspace(1,5,50)
plt.hist(x,bins=bin,ec='black')
plt.show()
```

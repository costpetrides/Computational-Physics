```python
import numpy as np
A = np.array([ [1. ,2. , 1. ], 
               [2. ,2. , 3. ], 
               [-1., -3., 0.] ])

(n,m)=A.shape
U=A.copy()

L=np.identity(n)
for k in range(n-1):
  for j in range(k+1,n):
    L[j,k]=U[j,k]/U[k,k]
    U[j,k:n]=U[j,k:n]-L[j,k]*U[k,k:n]

print('L matrix is \n', L)
print('')
print('U matrix is \n', U)
```

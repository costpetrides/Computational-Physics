```python
import matplotlib.pyplot as plt
import numpy as np
import math

x= [-1.6, -1.0, -0.4, 0.2, 0.8, 1.4]
y= [0.278037, 0.606531, 0.923116, 0.980199, 0.726149, 0.375311]
n=len(x)

yp = 0  # By default
xp= 1   # interpolation point


# Implementing Lagrange Interpolation
for k in range(n):
    L = 1
    for i in range(n):
        if i != k:
            L = L * (xp - x[i])/(x[k] - x[i])
    yp = yp + L * y[k]


print('Interpolated value at %.5f is %.5f.' % (xp, yp))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([-1.6, -1.0, -0.4, 0.2, 0.8, 1.4])
y= np.array([0.278037, 0.606531, 0.923116, 0.980199, 0.726149, 0.375311])

xplt=np.linspace(x[0],x[-1])
yplt=np.array([],float)

for xp in xplt:
  yp=0

  for xi,yi in zip(x,y):
    yp+= yi * np.prod((xp- x[x != xi])/(xi- x[x != xi]))
  yplt=np.append(yplt,yp)

plt.plot(x,y,'ro',xplt,yplt,'b-')
plt.show()
```

# 8.2

```python
x = [-1.6, -1.0, -0.4, 0.2, 0.8, 1.4]
y = [0.278037, 0.606531, 0.923116, 0.980199, 0.726149, 0.375311]
n=len(x)

def Q(xo):
  Q = np.zeros((n,n))

  for i in range(n):
    Q[i][0]=y[i]
    for j in range(1, i+1):
      Q[i][j] = ((xo - x[i-j]) * Q[i][j-1] - (xo - x[i]) * Q[i-1][j-1]) / (x[i] - x[i-j])
  return Q[i][i]

print('Interpolated value for f(0)= %.4f' %(Q(0)))
```

```python
xplot = np.linspace(x[0],x[-1])
yplot = np.zeros(len(xplot))

i = 0
for k in xplot:
  yplot[i] += Q(k)
  i += 1

plt.plot(x,y,'ro',xplt,yplt,'b-')
plt.show()
```

```python
w=np.linspace(100000,1900000,10)
t=[0.0000001,0.000001,0.00001]
R=5000
L=0.004
I0=0.025
Ui=[]

for ti in range (len(t)):
  U=[(w[l]*R*L*I0)/(np.sqrt(R*R+w[l]*w[l]*L*L))*math.sin(w[l]*t[ti] + math.atan2(R,w[l]*L)) for l in range (len(w))]
  Ui.append(U)
  print(" --------------")
  print("| For t=",t[ti],"|")
  print(" --------------")
  print("omega(Hz)\t\t\t\t\tU(V)")
  print()
  for x,y in zip(w,U):
    print(x,"\t\t\t\t",y)

  plt.plot(w,U)
  plt.xlabel("$\omega$ (MHz)")
  plt.ylabel("U (Volts)")
  plt.text(0.25*pow(10,6),70,t[ti])
  plt.show()
```

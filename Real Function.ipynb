**1. y=f(x)**


Example: $y=e^x  \;\;, \;\;[0,10]$

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)

def f(x):
  return np.exp(x)

y=[f(i) for i in x]

plt.plot(x,y, label='y=e^x')    # 2D Graph
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**2. z=f(x,y)**

Example: $z=x^2 \;+\;y^2  $

```python
import numpy as np                  
import matplotlib.pyplot as plt


def f(x, y):                         # Ορίζω την συνάρτησn f(x,y) = x^2 + y^2           
    return (x ** 2 + y ** 2)

x = np.linspace(-10, 10, 100)        # Θέτω τον πίνακα x = [-10,10] με 100 στοιχεία
y = np.linspace(-10, 10, 100)        # Θέτω τον πίνακα y = [-10,10] με 100 στοιχεία

X, Y = np.meshgrid(x, y)             # Ορίζώ ορθογώνιο πλέγμα από τους  δύο πίνακες x και  y

Z = f(X, Y)                          # Καλώ την συνάρτηση ως Z

fig = plt.figure()                   # 3D Γραφικά 
ax = plt.axes(projection="3d")   

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,      # Κατασκευή Γραφικής 
                cmap='winter', edgecolor='none')

ax.set_title('Title');             # Τίτλος
ax.set_xlabel('x')                 # Τίτλος  x - Άξονα
ax.set_ylabel('y')                 # Τίτλος  y - Άξονα
ax.set_zlabel('z')                 # Τίτλος  z - Άξονα

plt.show()
```

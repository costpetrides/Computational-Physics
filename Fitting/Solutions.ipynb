```python
import numpy as np
import matplotlib.pyplot as plt


x=np.array([1,2,3,4,5])
y=np.array([2.9,4.1,5.1,5.8,7.2])

S0=5
S1=np.sum(x)
S2=np.sum(x**2)
M0=np.sum(y)
M1=np.sum(x*y)

b=((M0/S0)-(M1/S1))/((S1/S0)-(S2/S1))
a=(M1-b*S2)/S1


def yy(x):
  return b*x+a

xx=np.linspace(x[0],x[-1],100)

plt.scatter(x,y,c='red')
plt.plot(xx,yy(xx),label='y=%.3f+%.3f' %(b,a))
plt.legend()
plt.show()
```

```python
A=np.array([[3.,  3., 5.],
            [3.,  5., 9.],
            [5., 9., 17.]])

b=np.array([[2.],
            [2.],
            [4.]])

C=np.concatenate((A,b),axis=1)
(n,m)=C.shape

for i in range(n):
  for j in range(n):
    if i!=j:
      ratio=C[j][i]/C[i][i]
      for k in range(m):
        C[j][k]=C[j][k] - ratio*C[i][k]


x=np.zeros(n)

for i in range(n):
  x[i]=C[i][n]/C[i][i]

for i in range(n):
    print('x[%.f]=%.2f' %(i+1,x[i]))
```

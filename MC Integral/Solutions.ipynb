```python
import numpy as np
import matplotlib.pyplot as plt
import math
def u(): return np.random.rand()
```

# 6.1

**i. Hit or Miss**

```python
def f(x): return 1/x

a, b = 0.5, 10.5
N    = 10**4

fmax = f(a)
V=fmax*(b-a)

hit=0
xhit=[]
yhit=[]
for i in range(N):
  x=(b-a)*u()+a
  y=fmax*u()
  if y<=f(x):
    hit+=1
    xhit+=[x]    # Graph
    yhit+=[y]    # Graph

p  = hit/N
I  = V*p
δI = V*(p-p*p)**0.5/N**0.5
print('hit-or-miss MC: I = %2.3f ± %2.3f for N = %d'%(I, δI, N))
```

**ii. Crude 1 dim**

```python
N   = 10**4
Σf  = 0
Σf2 = 0
for i in range(int(N)):
    x = u()*(b-a) + a
    Σf  += f(x)
    Σf2 += f(x)*f(x)


μf  = Σf/N
σf2 = Σf2/N - μf**2
sf2 = σf2*(N/(N-1))
sf  = sf2**0.5

V   = b - a
I   = V*μf
δI  = V*sf/N**0.5

print('crude MC: I = %2.3f ± %2.3f for N = %d'%(I, δI, N))
```

```python
# Graph
plt.scatter(xhit,yhit,c='green')
plt.show()
```

# 6.3

**i. Hit or Miss**


```python
def f(x,y):
  return (20/13)*(x+y)

fmax=f(5,1)
xa=0
xb=5
ya=0
yb=1

N=10**4
V=(xb-xa)*(yb-ya)*fmax

hit=0
xhit=[]  # Graph
yhit=[]  # Graph
zhit=[]  # Graph

for i in range(N):
  x=(xb-xa)*u() +xa
  y=(yb-ya)*u() +ya
  w=fmax*u()
  if w<=f(x,y):
    hit+=1
    xhit+=[x]    # Graph
    yhit+=[y]    # Graph
    zhit+=[w]    # Graph


p  = hit/N
I  = V*p
δI = V*(p-p*p)**0.5/N**0.5
print('hit-or-miss MC: I = %2.3f ± %2.3f for N = %d'%(I, δI, N))
```

**ii. Crude n dim**

```python
def f(x,y):
  return (20/13)*(x+y)

xa=0
xb=5
ya=0
yb=1
Sf=0
Sf2=0

V=5
N=10**4

for i in range(N):
  x=(xb-xa)*u() +xa
  y=(yb-ya)*u() +ya
  if 0<=x<=5 and 0<=y<=1:
    Sf+=f(x,y)
    Sf2 += f(x,y)*f(x,y)

μf  = Sf/N
σf2 = Sf2/N - μf**2
sf2 = σf2*(N/(N-1))
sf  = sf2**0.5

I   = V*μf
δI  = V*sf/N**0.5

print('crude MC: I = %2.3f ± %2.3f for N = %d'%(I, δI, N))
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(xhit, yhit, zhit, color = "green")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

nos = np.random.uniform(0, 1, int(1e4))
fcn = lambda x : (-65536*x**8 + 262144*x**7 - 409600*x**6 + 311296*x**5 -\
                  114688*x**4 + 16384*x**3)/27
```

```python
xx = np.arange(0, 1, 0.001)
plt.plot(xx, fcn(xx), 'k-', linewidth = 1.2)
plt.xlabel('x')
plt.ylabel('f(x)');
```

```python
n = int(1e4)
hits = 0
misses = 0
xH = []
yH = []
xM = []
yM = []
for i in range(n):
    xx = np.random.uniform()
    yy = np.random.uniform()
    if yy <= fcn(xx):
        hits += 1
        xH = xH + [xx]
        yH = yH + [yy]
    else:
        misses += 1
        xM = xM + [xx]
        yM = yM + [yy]
intEst = hits/n



plt.plot(xH, yH, 'go', fillstyle = 'none', markersize = 1.5)
plt.plot(xM, yM, 'ro', fillstyle = 'none', markersize = 1.5)
xx = np.arange(0, 1, 0.001)
plt.plot(xx, fcn(xx), 'k-', linewidth = 3)
plt.xlabel('x')
plt.ylabel('f(x)');
```

```python
import numpy as np
def U(a,b): return (b-a)*np.random.uniform(0,1) + a
def rho(x,y): return (20/13.)*(x+y)
n     = 1000
hit    = 0


for i in range(n):
    x = U(0,1)
    y = U(0,1)
    z = U(0,2*20/13.) 
    if x >= 0 and y <= 1 and y >= (x*x):
      if z <= rho(x,y): 
        hit+= 1 


V  = 2*(20/13.)
p  = hit/n
I  = V*hit/n
dI = V*(p-p*p)**0.5/n**0.5
print('I = %2.4f +/- %2.4f'%(I,dI))
```

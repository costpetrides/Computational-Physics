```python
import numpy as np 
import matplotlib.pyplot as plt
from math import *

def f(x):
  return 7*x*x

def u():
  return np.random.rand()

fmax=f(3)
V=fmax*2

hit=0
N=1000

for i in range(N):
  x=2*u()+1
  y=fmax*u()
  if y<=f(x):
    hit+=1

P=hit/N

I=P*V
I=V*np.sqrt(P-P**2)/np.sqrt(N)

print('Integral I= %.3f +/- %.3f' %(I,I))
```

Lets play a game! We will calculte the Integrl 1000 times. Is the result 60.666 ?

```python
def game():
  hit=0
  N=1000
  for i in range(N):
    x=2*u()+1
    y=fmax*u()
    if y<=f(x):
      hit+=1
  P=hit/N
  I=P*V
  return I

data=[game() for i in range(1000)]
plt.hist(data,bins=np.linspace(50,70,10),ec='black')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from math import *


def g(x,y):
  c=20/13
  return c*(x+y)

def U(a,b):
  return (b-a)*np.random.rand() + a

gmax=g(5,1)
V=gmax*(5-0)*(1-0)

hit=0
N=1000

for i in range(N):
  x=U(0,5)
  y=U(0,1)
  z=U(0,gmax)
  if z<=g(x,y):
    hit+=1

P=hit/N

I=P*V
I=V*np.sqrt(P-P**2)/np.sqrt(N)

print('Integral I= %.3f +/- %.3f' %(I,I))
```

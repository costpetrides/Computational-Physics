```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 7*x**2

def u(a,b):
  return (b-a)*np.random.rand()+a

a=1
b=3
N   = 10**4
f  = 0
f2 = 0

for i in range(N):
    x = u(a,b)
    f  += f(x)
    f2 += f(x)*f(x)
    
f  = f/N
f2 = f2/N - f**2
sf2 = f2*(N/(N-1))
sf  = sf2**0.5

V   = b - a
I   = V*f
I  = V*sf/N**0.5

print('crude MC: I = %2.3f  %2.3f for N = %d'%(I, I, N)) 
```

**Or we can use arrays...!!**


```python
F=[]
for i in range(N):
  x = u(a,b)
  F+=[f(x)]

F=np.array(F)

I=np.mean(F)*V
sf=np.std(F)
I  = V*sf/N**0.5

print('crude MC: I = %2.3f  %2.3f for N = %d'%(I, I, N)) 
```

```python
def f(x,y):
  return (20/13)*(x+y)

def u(a,b):
  return (b-a)*np.random.rand() +a
xa=0
xb=5
ya=0
yb=1
Sf=0
Sf2=0

V=5
N=10**4

for i in range(N):
  x=u(xa,xb)
  y=u(ya,yb)
  Sf+=f(x,y)
  Sf2 += f(x,y)*f(x,y)

f  = Sf/N
f2 = Sf2/N - f**2
sf2 = f2*(N/(N-1))
sf  = sf2**0.5

I   = V*f
I  = V*sf/N**0.5

print('crude MC: I = %2.3f  %2.3f for N = %d'%(I, I, N)) 
```

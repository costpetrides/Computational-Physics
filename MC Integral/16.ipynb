```python
import numpy as np
import matplotlib.pyplot as plt
import math
def u(): return np.random.rand()

def f(x): return (np.sin(np.sqrt(x)+1) * np.exp(np.sqrt(x))) / np.sqrt(x)

a, b = 1, 4
N    = 10**4

fmax = f(a)
V=fmax*(b-a)

hit=0
xhit=[]
yhit=[]
for i in range(N):
  x=(b-a)*u()+a
  y=fmax*u()
  if y<=f(x):
    hit+=1
    xhit+=[x]    # Graph
    yhit+=[y]    # Graph

p  = hit/N
I  = V*p
δI = V*(p-p*p)**0.5/N**0.5
print('hit-or-miss MC: I = %2.3f ± %2.3f for N = %d'%(I, δI, N))


xx=np.linspace(a,b,1000)
plt.plot(xx,f(xx),linewidth='2',color='red')
plt.scatter(xhit,yhit,c='green')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u():
    return np.random.rand()

r = 5
N = 10**4

V = (2*r)**3  # Όγκος κύβου πλευράς 2r

hit = 0
xhit, yhit, zhit = [], [], []

for _ in range(N):
    x = 2*r*u() - r  # τυχαίο σημείο στο [-r, r]
    y = 2*r*u() - r
    z = 2*r*u() - r

    # Έλεγχος αν το σημείο είναι εκτός της σφαίρας (οπή)
    if x**2 + y**2 + z**2 >= r**2:
        hit += 1
        xhit.append(x)
        yhit.append(y)
        zhit.append(z)

p = hit / N
I = V * p
delta_I = V * np.sqrt(p * (1-p)) / np.sqrt(N)

print(f'Hit-or-miss MC: Όγκος = {I:.5f} ± {delta_I:.5f} για N = {N}')

# Οπτικοποίηση των σημείων που μετρήθηκαν (εκτός σφαίρας)
fig = plt.figure(figsize=(10, 7))
ax = plt.axes(projection='3d')
ax.scatter3D(xhit, yhit, zhit, color='green', s=1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Monte Carlo Hit or Miss: Κύβος με σφαιρική οπή')
plt.show()
```

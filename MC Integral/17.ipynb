```python
import numpy as np
import matplotlib.pyplot as plt
import math

def u():
  return np.random.rand()

r=5
N    = 10**4

V=(2*r)**3

hit=0
xhit=[]
yhit=[]
zhit=[]

for i in range(N):
  x=2*r*u() - r  # Unif(-r,+r)
  y=2*r*u() - r
  z=2*r*u() - r
  if x**2 + y**2 + z**2 >= r**2:
    hit+=1
    xhit+=[x]    # Graph
    yhit+=[y]    # Graph
    zhit+=[z]    # Graph

p  = hit/N
I  = V*p
Î´I = V*(p-p*p)**0.5/N**0.5
print('hit-or-miss MC: I = %2.3f Â± %2.3f for N = %d'%(I, Î´I, N))

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(xhit, yhit, zhit, color = "green")
plt.show()
```

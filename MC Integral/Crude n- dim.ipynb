```python
import numpy as np 


def u(a,b): 
  return (b-a)*np.random.uniform(0,1) + a

def pd(x,y):
  return  (20/13.)*(x+y)  




repeat=10000       
N=0        
f=0       # Sum to calculate the integral
f2=0      # Sum is to calculate the integrals standar diviation

for i in range(repeat):
  x=u(0,1)
  y=u(0,1)
  f=pd(x,y)
  if (x>=0 and x**2<=y and y<=1):
    N+=1
    f+=f
    f2+=f**2

mean=f/N
sigmaf2=(f2/N-mean**2)*N/(N-1)
sigmaf=sigmaf2**0.5

Volume=2/3.    #Volume of domain
I=mean*Volume
dI=Volume*sigmaf/ N**0.5


print('I = %2.4f +/- %2.4f'%(I,dI))
```

```python
def u(a,b):
  return (b-a)*np.random.uniform(0,1) + a

def p(x,y,z):
  return (5/(648*np.pi))*(x**2 + y**2)

m=0
repeat=10000
f=0
f2=0
V=(4/3.)*np.pi*3**3

for i in range(repeat):
  x=u(-3,3)
  y=u(-3,3)
  z=u(-3,3)
  f=p(x,y,z)
  if x**2 + y**2 + z**2<=3**2:
    f+=f
    f2+=f**2
    m+=1

mean=f/m
sf2= f2/m-mean**2
sigmaf2=(m/(m-1))*sf2
sf=sigmaf2**0.5

I=mean*V
dI=(V/(m**0.5))*sf

print('I = %2.4f +/- %2.4f'%(I,dI))
```

**Bisection method**
bisection.PNG

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return pow(x,3)-x-1
a = 1
b = 2
n = 0
e = 1.e-6 # = 0.00001
while( b - a > e and n < 1000):
  c = 0.5*(a + b)
  if f(c)*f(b) < 0: a = c
  else: b = c
  n = n + 1
  print('n = ', n, 'a = ', a, 'b = ', b)

po=np.linspace(1,2,1000)
plt.plot(po,f(po))
plt.axhline(color='black')
```

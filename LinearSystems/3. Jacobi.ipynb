```python
j1,j2,j3=1,2,3
ep=1.e-6

for i in range(10):
  x1 = (10 + j2)/3.
  x2 = (5 + j1 + j3)/3.
  x3 = (10 + 2*j2)/4.
  j1,j2,j3=x1,x2,x3
  if (abs(x1-j1)>ep or abs(x2-j2)>ep or abs(x3-j3)>ep):
    break
  
print('X[1]= %.3f' %(x1))
print('X[2]= %.3f' %(x2))
print('X[3]= %.3f' %(x3))
print('N=%.f' %(i+1))
```

```python
from numpy.matrixlib.defmatrix import concatenate
import numpy as np
import matplotlib.pyplot as plt
import sys

A=np.array([[ 1., 2., 1.],
            [ 2.,-2., 1.],
            [-1.,-3., 0.]])

b=np.array([[0.],
            [3.],
            [2.]])

C=np.concatenate((A,b),axis=1)
(n,m)=C.shape

ratio=0

#Gauss Elimination
for i in range(n):
  for j in range(n):
    if i!=j:
      ratio=C[j][i]/C[i][i]
      for k in range(m):
        C[j][k]=C[j][k]-ratio*C[i][k]


#Obtaining Solutions

x=np.zeros(n)
for i in range(n):
  x[i]=C[i][n]/C[i][i]

for i in range(n):
    print('x[%.f]=%.2f' %(i+1,x[i]))
```

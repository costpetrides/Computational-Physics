```python
x1,x2,x3=1.,2.,3.

for i in range(10):
    x1 = (10 + x1)/3.
    x2 = (5 + x1 + x3)/3.
    x3 = (10 + 2*x2)/4.

print('X[1]=%.3f' %(x1))
print('X[2]=%.3f' %(x2))
print('X[3]=%.3f' %(x3))
```

**i. Gauss Elimination**

```python
import numpy as np

# Define the matrix A and vector b
A = np.array([[2., 1., 1.],
              [2., -1., 2.],
              [1., -2., 1.]])

b = np.array([[7.],
              [6.],
              [0.]])

# Concatenate A and b to form augmented matrix C
C = np.concatenate((A, b), axis=1)
(n, m) = C.shape

# Forward elimination
for i in range(n):
    # Make the diagonal contain all 1's and others 0's !
    # Divide the pivot row by the pivot element
    C[i] = C[i] / C[i, i]

    # Make the elements below the pivot equal to 0
    for j in range(i + 1, n):
        ratio = C[j, i]
        C[j] = C[j] - ratio * C[i]

# Back substitution
x = np.zeros(n)
for i in range(n - 1, -1, -1):
    x[i] = C[i, -1] - np.sum(C[i, i+1:n] * x[i+1:n])

# Output the results
for i in range(n):
    print(f'x[{i+1}] = {x[i]:.2f}')
```

**ii. L-U U method (Doolittle)**

```python
import numpy as np

# Define the matrix A
A = np.array([[2., 1., 1.],
              [2., -1., 2.],
              [1., -2., 1.]])

# Initialize L as the identity matrix and U as a copy of A
n = A.shape[0]
U = A.copy()
L = np.identity(n)

# Perform LU Decomposition
for k in range(n - 1):
    for j in range(k + 1, n):
        # Compute the multiplier
        L[j, k] = U[j, k] / U[k, k]

        # Update the U matrix
        U[j, k:n] = U[j, k:n] - L[j, k] * U[k, k:n]

# Print the results
print('L matrix:')
print(L)

print()

print('U matrix:')
print(U)
```

```python
C=np.concatenate((L,b),axis=1)
(n,m)=C.shape

for i in range(n):
  for j in range(n):
    if i!=j:
      ratio=C[j][i]/C[i][i]
      for k in range(m):
        C[j][k]=C[j][k] - ratio*C[i][k]

y=np.zeros(n)

for i in range(n):
  y[i]=C[i][n]/C[i][i]

for i in range(n):
    print('y[%.f]=%.2f' %(i+1,y[i]))
```

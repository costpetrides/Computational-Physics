```python
import numpy as np
import sys

A = np.array([ [1. ,2. , 1. ], 
               [2. ,2. , 3. ], 
               [-1., -3., 0.] ])

b = np.array([ [0.],   # note the format of np.array([ ])
               [3.],   # np.array([ [0], [3], [2] ]) has shape 3 x 1 
               [2.] ]) # while np.array([[ 0, 3, 2 ]) would have 1 x 3 


C = np.concatenate((A,b), axis = 1)

(n,m)=C.shape
x = np.zeros(n)



# Applying Gauss Jordan Elimination
for i in range(n):
 for j in range(n):
    if i != j:
      ratio = C[j][i]/C[i][i]
      for k in range(m):
        C[j][k] = C[j][k] - ratio * C[i][k]

# Obtaining Solution

for i in range(n):
    x[i] = C[i][n]/C[i][i]

# Displaying solution
print('\nThe solution is: ')

for i in range(n):
    print(x[i])
```

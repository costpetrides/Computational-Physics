Solve the ODE for τ=0.02:

$y'=0.5y , \: \:\: \: \: 0\leq t \leq 2 \: \: ,\: \: \: y(0)=1$ 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,y):
  return 0.5*y

a=0
b=2
c=1

τ=0.02
N=np.ceil((b-a)/τ)
n=int(N)


w=np.zeros(n+1)
t=np.zeros(n+1)

w[0]=c
t[0]=0

for i in range(n):
  w[i+1]=w[i] + τ*f(t[i],w[i])
  t[i+1]=t[i] + τ

print('The solution for τ=%.2f is %.3f' %(τ,w[-1]))
```

**Graph for different τ**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,y):
  return 0.5*y

def sol(x):
  a=0
  b=2
  c=1

  τ=x
  N=np.ceil((b-a)/τ)
  n=int(N)


  w=np.zeros(n+1)
  t=np.zeros(n+1)

  w[0]=c
  t[0]=0

  for i in range(n):
    w[i+1]=w[i] + τ*f(t[i],w[i])
    t[i+1]=t[i] + τ
  
  return w

plt.plot(sol(0.5),label='τ=0.6')
plt.plot(sol(0.2),label='τ=0.2')
plt.plot(sol(0.02),label='τ=0.09')
plt.plot(sol(0.002),label='τ=0.02')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Computational-Physics/main/LIDAR/K532.csv'
df = pd.read_csv(url)
print(df)
```

# Backscatter b

```python
plt.figure(figsize=(4,4))

plt.xlabel('Î² [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Height [m]')

plt.plot(df['LR120'][85:3993], df['Z'][85:3993], 'cyan', label='LR=120')

plt.plot(df['LR100'][85:3993], df['Z'][85:3993], 'yellow', label='LR=100')

plt.plot(df['LR80'][85:3993], df['Z'][85:3993], 'orange', label='LR=80')

plt.plot(df['LR60'][85:3993], df['Z'][85:3993], 'blue', label='LR=60')

plt.plot(df['LR40'][85:3993], df['Z'][85:3993], 'red', label='LR=40')

plt.plot(df['LR20'][85:3993], df['Z'][85:3993], 'green', label='LR=20')

plt.ylim(0,5000)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Relative change [%]')
plt.ylabel('Height [m]')

c120= (df['LR120'][85:3993] - df['LR60'][85:3993])/ df['LR60'][85:3993]
c100=(df['LR100'][85:3993] - df['LR60'][85:3993])/ df['LR60'][85:3993]
c80 =(df['LR80'][85:3993] - df['LR60'][85:3993])/ df['LR60'][85:3993]
c60 =(df['LR60'][85:3993] - df['LR60'][85:3993])/ df['LR60'][85:3993]
c40 =(df['LR40'][85:3993] - df['LR60'][85:3993])/ df['LR60'][85:3993]
c20 =(df['LR20'][85:3993] - df['LR60'][85:3993])/ df['LR60'][85:3993]

plt.plot(c120*100, df['Z'][85:3993], 'cyan', label='LR=120')

plt.plot(c100*100, df['Z'][85:3993], 'yellow', label='LR=100')

plt.plot(c80*100, df['Z'][85:3993], 'orange', label='LR=80')

plt.plot(c60*100, df['Z'][85:3993], 'blue', label='LR60')

plt.plot(c40*100, df['Z'][85:3993], 'red', label='LR40')

plt.plot(c20*100, df['Z'][85:3993], 'green', label='LR20')

plt.ylim(0,4300)
plt.xlim(-220,220)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Absolute change [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Hight [m]')

c120= abs(df['LR120'][85:3993] - df['LR60'][85:3993])
c100=abs(df['LR100'][85:3993] - df['LR60'][85:3993])
c80 =abs(df['LR80'][85:3993] - df['LR60'][85:3993])
c60 =abs(df['LR60'][85:3993] - df['LR60'][85:3993])
c40 =abs(df['LR40'][85:3993] - df['LR60'][85:3993])
c20 =abs(df['LR20'][85:3993] - df['LR60'][85:3993])

plt.plot(c120*100, df['Z'][85:3993], 'cyan', label='LR=120')

plt.plot(c100*100, df['Z'][85:3993], 'yellow', label='LR=100')

plt.plot(c80*100, df['Z'][85:3993], 'orange', label='LR=80')

plt.plot(c60*100, df['Z'][85:3993], 'blue', label='LR=60')

plt.plot(c40*100, df['Z'][85:3993], 'red', label='LR=40')

plt.plot(c20*100, df['Z'][85:3993], 'green', label='LR=20')

plt.ylim(0,4500)
plt.legend()

plt.show()
```

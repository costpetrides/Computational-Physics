**Create radnom numbers (0,1)**

```python
import numpy as np
import matplotlib.pyplot as plt
x=[]
def u():
  return np.random.rand()            #function that gives one radnom  number (0,1)

x=[np.sqrt(u()) for i in range(0,1000)]         # Tips!!
plt.hist(x,bins=10, ec='black')
```

**Solve the ODE for =0.5:**

**$y'=0.5y , \: \:\: \: \: 0\leq t \leq 2 \: \: ,\: \: \: y(0)=1$**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,y):
  return 0.5*y

a=0
b=2
c=1
=0.5

n=int((b-a)/)

w=np.zeros(n+1)
t=np.zeros(n+1)

w[0]=c
t[0]=0


for i in range(n):
  w[i+1]=w[i]+*f(t[i]+(/2),w[i]+(/2)*f(t[i],w[i]))
  t[i+1]=t[i]+1

print('The solution fr =0.5 (2nd Runge Kutta) is  %.3f '%(w[-1]))
```

**Solve the ODE for =0.5:**

**$y'=0.5y , \: \:\: \: \: 0\leq t \leq 2 \: \: ,\: \: \: y(0)=1$**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,y):
  return 0.5*y

a=0
b=2
c=1
=0.5

n=int((b-a)/)

w=np.zeros(n+1)
t=np.zeros(n+1)

w[0]=c
t[0]=0


for i in range(n):
  k1=*f(t[i],w[i])
  k2=*f(t[i]+/2,w[i]+k1/2)
  k3=*f(t[i]+/2,w[i]+k2/2)
  k4=*f(t[i]+,w[i]+k3)
  w[i+1]=w[i]+(1/6)*(k1+2*k2+2*k3+k4)


print('The solution fr =0.5 (4nd Runge Kutta) is  %.3f '%(w[-1]))
```

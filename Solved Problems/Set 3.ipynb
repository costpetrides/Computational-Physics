```python
# 6a,1)

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.tan(x) - x/(1-x**2)

a=2
b=4

for i in range(15):
  c=(a+b)/2
  if f(a)*f(b)<0:
    b=c
  if f(b)*f(c)<0:
    a=c

print('The root is : Xo= %.3f' %(c))

print('f(%.3f)=%.3f' %(c,f(c)))
```

```python
# 6a,2)

def df(x):
  return (1/np.cos(x))**2 - 1/(1-x**2) - (2*x**2)/(1-x**2)**2

n=6
x=3.0
for i in range(n):
  x=x-f(x)/df(x)

print('The root is : Xo= %.3f' %(x))
print('f(%.3f)=%.3f' %(x,f(x)))
```

```python
# 6b)

def bisec(a,b):

  for i in range(100):
    c=(a+b)/2
    if abs(f(c))<=1.e-6:
      break
    if f(a)*f(b)<0:
      b=c
    if f(b)*f(c)<0:
      a=c
    


  print('Root in [a,b]')
  print('The root is : Xo= %.3f' %(c))
  print('f(%.3f)=%.3f' %(c,f(c)))
  print('\n')

bisec(4,5)
bisec(5,6)
bisec(6,7)
bisec(7,8)
bisec(8,9)
bisec(9,10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gsiedel(n):

  x,y,z=0,0,0

  for i in range(n):
    x=(3-2*y-3*z)/2
    y=(2+x)/(-3)
    z=-(x+2*y)

  print('For n= %.f  X=[%.3f, %.3f, %.3f]'  %(n,x,y,z))

gsiedel(3)
gsiedel(200)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Jacobi(n):

  j1,j2,j3=0,0,0

  for i in range(n):
    x1=(3-2*j2-3*j3)/2
    x2=(2+j1)/(-3)
    x3=-(j1+2*j2)
    j1=x1
    j2=x2
    j3=x3

  print('For n= %.f  X=[%.3f, %.3f, %.3f]'  %(n,j1,j2,j3))

Jacobi(3)
Jacobi(200)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u():
  x=6*np.random.rand() + 1
  return int(x)

def game():
  pos=1
  n=0
  while pos<25:
    n+=1
    pos+=u()
    if pos==3:
      pos=24
    if pos==21:
      pos=7
  return n
  
dice=[game() for i in range(200)]
plt.hist(dice,bins=range(0,30),ec='black')
plt.xlabel('dice to win')
```

```python
# 3a,b)

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.exp(x)

def u():
  return 10*np.random.rand()

n=1000
f=0
f2=0
for i in range(n):
  x=u()
  y=f(x)
  f+=y
  f2+=y**2

mean=f/n
I=mean*10

sf2=(f2/n)-(mean**2)
std=np.sqrt(sf2)

I=10/np.sqrt(n)*std

print('Integral for n=%.f : %3.f +/- %.3f ' %(n,I,I))
print('/= %.3f' %(I/I))
```

```python
def un(a,b):
  return (b-a)*np.random.rand() + a

def In(a,b):
  n=500
  f=0
  
  for i in range(n):
      x=un(a,b)
      y=f(x)
      f+=y

  mean=f/n

  I=mean*5

  return I

Integer=In(0,5) + In(5,10)
print('Integer value is : %3.f' %(Integer))
```

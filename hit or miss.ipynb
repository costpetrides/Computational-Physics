```python
import numpy as np
import matplotlib.pyplot as plt

def u(): return np.random.rand()

def f(x): return np.exp(x)/(np.exp(5)-1)        #PDF 
xhit=[] 
xmiss=[]
fmax=f(5)
for i in range(10000):
  x=5*u()
  y=fmax*u()
  
  if y<= f(x):
    xhit+=[x]
  else:
    xmiss+=[x]

print(xhit)

plt.hist(xhit, bins=100, alpha=0.5, label='strike!')
plt.hist(xmiss, bins=100, alpha=0.5, label='miss')
plt.legend(loc='upper right')
plt.show()

plt.hist(xhit,bins=100,ec='black')
plt.show()
```

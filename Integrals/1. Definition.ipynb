$I= \int_{0}^{4} e^{x}+ 5x^{2} dx$

```
import numpy as np

def f(x):
  return np.exp(x) + 5*x**2

a=0
b=4


I=0
x=np.linspace(a,b,1000)
h=(b-a)/len(x)
for xi in x:
  I+=h*f(xi)

print("I=",I)
```

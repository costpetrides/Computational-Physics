```python
import numpy as np

def f(x):
  return np.exp(x)+5*x**2

a=0
b=1

I=0
while b<=4:
  h=(b-a)/(2)
  I+=(h/3)*(f(a)+4*f(h+a)+f(b))
  a+=1
  b+=1


print('The integral value is I=%.3f' %(I))
```

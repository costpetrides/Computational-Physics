```python
import numpy as np

def f(x):
  return np.exp(x)+5*x**2
a=0
b=1
h=(b-a)
I=0

while (b<=4):
  I+=(h/2)*(f(a)+f(b))
  a+=1
  b+=1

print('The integral value is I=%.3f' %(I))
```

Computational Phys (7).png

```python
import numpy as np

def f(x):
  return np.exp(x)+5*x**2

a=0
b=4


x=np.linspace(a,b,100)
n=len(x)
h=(b-a)/n
s=np.sum(2*f(x))
I=(h/2)*(f(a)+s+f(b))

print('The integral value is I=%.3f' %(I))
print(n)
```
